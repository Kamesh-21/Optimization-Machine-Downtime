### Replacing Null Values with Mean & Median

```python
# Define columns to replace NaN with mean
mean_columns = ['Hydraulic_Pressure(bar)', 'Coolant_Pressure(bar)', 'Air_System_Pressure(bar)',
                'Coolant_Temperature', 'Hydraulic_Oil_Temperature(°C)', 'Spindle_Bearing_Temperature(°C)',
                'Spindle_Vibration(µm)', 'Tool_Vibration(µm)', 'Torque(Nm)', 'Cutting(kN)']

# Define columns to replace NaN with median
median_columns = ['Spindle_Speed(RPM)', 'Voltage(volts)']

# Fill NaN values with mean for selected columns
dataset[mean_columns] = dataset[mean_columns].apply(lambda x: x.fillna(x.mean()))

# Fill NaN values with median for selected columns
dataset[median_columns] = dataset[median_columns].apply(lambda x: x.fillna(x.median()))
```

##### Skewness is a statistical measure that indicates the asymmetry of a data distribution.
##### Positive Skew (Right-Skewed) → Data is stretched towards the right (Mean > Median > Mode).
##### Negative Skew (Left-Skewed) → Data is stretched towards the left (Mean < Median < Mode).
##### Zero Skew → Data is symmetric.

##### Kurtosis measures the peakedness (fatness of tails) of a data distribution.
##### High Kurtosis (Leptokurtic, >3) → The data has heavy tails (many outliers).
##### Low Kurtosis (Platykurtic, <3) → The data is flat with fewer outliers.
##### Normal Kurtosis (Mesokurtic, ≈3) → The data follows a normal distribution.

```python
#### Monthly Trend of Machine Failures

# Convert 'Date' column to datetime format 
dataset['Date'] = pd.to_datetime(dataset['Date'], dayfirst=True)
# Extract month and year
dataset["Month_Year"] = dataset["Date"].dt.to_period("M")

# Filter only Machine Failure cases
machine_failure_data = dataset[dataset["Downtime"] == "Machine_Failure"]

# Count failures per month
monthly_failures = machine_failure_data.groupby("Month_Year").size()

# Plot Line Chart
plt.figure(figsize=(6,3))
plt.plot(monthly_failures.index.astype(str), monthly_failures.values, marker='o', linestyle='-', color='b')

# Formatting
plt.xticks(rotation=45, fontsize=8)
plt.yticks(fontsize=9)
plt.xlabel("Month-Year", fontsize=10)
plt.ylabel("Number of Machine Failures", fontsize=8)
plt.title("Monthly Trend of Machine Failures", fontsize=10)
plt.grid(True)

# Show Plot
plt.show()
```

```python
### No_Machine_Failure v/s Machine_Failure Month-wise

# Extract the month name
dataset['Month'] = dataset['Date'].dt.strftime('%B')

# Group by Month and Downtime status
monthly_failures = dataset.groupby(['Month', 'Downtime']).size().unstack()

# Rename columns for better readability
monthly_failures.columns = ['No_Machine_Failure', 'Machine_Failure']

# Fill NaN values with 0 (if a category is missing in some months)
monthly_failures = monthly_failures.fillna(0).astype(int)

# Reorder months correctly
month_order = ["January", "February", "March", "April", "May", "June", "July", 
               "August", "September", "October", "November", "December"]
monthly_failures = monthly_failures.reindex(month_order)

# Display the result
print(monthly_failures)
```

##### Machine failures are significantly high in March (515), February (301), and April (247) compared to other months.
##

```python
#### Correlation Between Machine Parameters and Machine Failure

# Convert 'Downtime' column to numerical values (1 = Machine Failure, 0 = No Machine Failure)
dataset["Failure_Flag"] = dataset["Downtime"].apply(lambda x: 1 if x == "Machine_Failure" else 0)

# Select only machine parameters + Failure Flag for correlation
columns_to_analyze = [
    "Hydraulic_Pressure(bar)", "Coolant_Pressure(bar)", "Air_System_Pressure(bar)", 
    "Coolant_Temperature", "Hydraulic_Oil_Temperature(°C)", "Spindle_Bearing_Temperature(°C)", 
    "Spindle_Vibration(µm)", "Tool_Vibration(µm)", "Spindle_Speed(RPM)", 
    "Voltage(volts)", "Torque(Nm)", "Cutting(kN)", "Failure_Flag"
]

# Compute correlation matrix
corr_matrix = dataset[columns_to_analyze].corr()

# Plot Heatmap
plt.figure(figsize=(8.5,4.5))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Formatting
plt.title("Correlation Between Machine Parameters and Machine Failure", fontsize=12)
plt.xticks(rotation=45, ha="right", fontsize=9)
plt.yticks(fontsize=9)

# Show Plot
plt.show()
```

##### Hydraulic Pressure (-0.56) lower pressure may cause issues.
##### Cutting force (0.45) has a positive failure correlation, indicating high cutting loads increase downtime risks.
##### LowTorque (-0.41) might lead to failures
##### Voltage & Torque (-0.19): Low voltage could reduce torque, affecting machine efficiency.

###  Machine Downtime Root Cause Analysis

##### -- **Low Coolant Pressure** leads to machine overheating, increased spindle vibration, and eventual breakdown, causing unplanned downtime and production loss.
Recommendation - Set alerts for coolant pressure drop and Optimize coolant circulation to prevent overheating.

##### -- **High Vibration** impacts machine performance, reduces tool life and increases downtime as well as repair & maintenance costs..  
Recommendation -Implement vibration monitoring to detect abnormal trends early & Use shock-absorbing mounts to reduce excessive vibration.
Controlling vibration can enhance tool life, improve product quality, and minimize unplanned downtime

##### -- **Voltage(volts)** instability can cause machines to shut down unexpectedly.
Recommendation - Ensure stable power supply using voltage stabilizers.

##### -- **High** Coolant_Temperature, Hydraulic_Oil_Temperature(°C), and Spindle_Bearing_Temperature(°C) indicate overheating problems.
Recommendation - Regularly check coolant systems and apply temperature threshold-based shutdowns before failure occurs.

##### -- Hydraulic Pressure fluctuating significantly, can cause stress on hydraulic systems, leading to breakdown.
Recommendation -  Set up alerts for hydraulic pressure deviations beyond safe limits. and schedule servicing before it breaks down.

##### -- when machines may operate at suboptimal torque, spindle speed, or cutting force, leading to increased stress and potential failures.
Recommendation -- Maintain optimal torque levels, keep speed within the recommended range, and balance cutting parameters ,so that reduce wear and tear.

##### --  Avoid Expensive Emergency Repairs
by Predicting failures early and scheduling maintenance reduces these emergency costs.
